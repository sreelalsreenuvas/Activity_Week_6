# 2. Data Visualization

```python
plt.figure(figsize=(15,10))
Region=data['Region']
Population=data['Population in thousands (2017)']
plot=plt.bar(Region,Population,width=.6)
plt.title('Population v/s Region')
plt.xticks(rotation='90')
plt.xlabel('Region')
plt.ylabel('Population in thousands (2017)')
plt.grid()
```

# 4. a) GDP per capita (current US$)

```python
GDP=data.groupby('Region')['GDP per capita (current US$)'].mean()
m_GDP=pd.DataFrame(GDP).reset_index()
m_GDP
```

```python
Highest_GDP = m_GDP.nlargest(1,'GDP per capita (current US$)')
Highest_GDP
```

# 4. b) International trade: Imports (million US$)

```python
Trade = data.groupby('Region')['International trade: Imports (million US$)'].mean()
m_Trade = pd.DataFrame(Trade).reset_index()
m_Trade
```

```python
Highest_Trade = m_Trade.nlargest(1,'International trade: Imports (million US$)')
Highest_Trade
```
